```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/thiru/Downloads/fundamentals/fundamentals.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data = data.iloc[:,3:]
```

```python
data
```

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
print(scaled_data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 25)
pca.fit(scaled_data)
```

```python
X = pca.transform(scaled_data)
print(scaled_data.shape)
print(X.shape)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:, 0],X[:, 1],X[:, 2], c = data['Estimated Shares Outstanding'],cmap = 'plasma')
ax.set_xlabel('First Component')
ax.set_ylabel('Second Component')
ax.set_zlabel('Third Component')
plt.show()
```

```python
pca.components_
```

```python
pca.get_covariance()
```

```python
pca.get_precision()
```

```python
pca.explained_variance_
```

```python
from sklearn.model_selection import train_test_split

Y = data['Estimated Shares Outstanding']
X_train,X_test,y_train,y_test= train_test_split(X,Y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
regressor.score(X_test,y_test)
```

```python
plt.scatter(y_test,y_pred)
```

```python
new_column_index = [np.abs(pca.components_[i]).argmax() for i in range(25)]

print('new_column_index =',new_column_index)
```

```python
new_column_set = []

for i in new_column_index:

    feature = data.columns[i]

    new_column_set.append(feature)

   

print('new_column_set =',new_column_set)
```

```python
df = pd.DataFrame(X,columns = new_column_set)
```
