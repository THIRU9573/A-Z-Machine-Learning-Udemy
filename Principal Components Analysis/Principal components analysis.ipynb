# what is Principal Component Analysis:

Principal Component Analysis (PCA) is a statistical technique used for data reduction and dimensionality reduction. It is a popular technique in machine learning and data analysis. PCA aims to reduce the dimensionality of a dataset while retaining as much information as possible. In simple terms, it tries to find the most important features or variables that explain the most significant amount of variance in the data.

Here's a step-by-step explanation of how PCA works:

1.Standardize the data: PCA is sensitive to the scale of the variables, so it is essential to standardize the data before applying PCA. This means that each variable is scaled to have a mean of 0 and a standard deviation of 1.

2.Calculate the covariance matrix: The covariance matrix describes the relationships between the different variables in the data. It shows how much two variables are related to each other.

3.Calculate the eigenvectors and eigenvalues of the covariance matrix: Eigenvectors are vectors that do not change direction during a transformation, and eigenvalues represent the magnitude of the transformation along the eigenvectors. The eigenvectors and eigenvalues of the covariance matrix are calculated to determine the principal components.

4.Choose the principal components: The principal components are the eigenvectors with the highest eigenvalues. The number of principal components chosen determines how much variance is retained in the data. Typically, the first two or three principal components are selected, as they explain the majority of the variance in the data.

5.Calculate the new feature vectors: The new feature vectors are created by projecting the data onto the chosen principal components. This transformation reduces the dimensionality of the data while retaining as much information as possible.

6.Interpret the results: The new feature vectors can be used for further analysis, such as clustering, classification, or visualization. They represent the most important features or variables in the data and can provide insights into the underlying patterns and relationships in the data.

Overall, PCA is a powerful technique that can be used to reduce the dimensionality of large datasets while retaining as much information as possible. It is widely used in machine learning, data analysis, and other fields where data reduction is needed.

# PCA VISUALIZATION:

As we noticed before it is difficult to visualize high dimensional data.we can use PCA to find the first two principal components and visualize the data in this new, two dimensional space,
and single scatter.beforewe do this through we will need to scale our data so that each feature has a single unit varience

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
print(scaled_data)
```

```python
scaled_data.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(scaled_data)
```

# Now we can transform this data to first 2 principal components

```python
X = pca.transform(scaled_data)
print(scaled_data.shape)
print(X.shape)
```

# Plot these two dimensions Out!:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cancer['target'], cmap='plasma')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

# Interpreting the Components:

1.Unfortunately with this great power of dimensionality reduction, comes the  cost of Not being able to easily understand what these components represent

2.the Components correspond to combination of original features, the components themselves are stoerd as an attribute of the fitted PCA object

```python
pca.components_
```

->In this numpy array each row represent a principal componet, and each column realte back to the original features

```python
df_comp = pd.DataFrame(pca.components_,columns = cancer['feature_names'])
plt.figure(figsize =(12,6))
sns.heatmap(df_comp,cmap = 'plasma')
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer() 
cancer.keys()
```

```python
data = pd.DataFrame(cancer.data,columns = cancer.feature_names)
data
```

```python
print(cancer['DESCR'])
```

```python
print(cancer['target_names'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data1 = scaler.fit_transform(data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
pca.fit(scaled_data1)
```

```python
X = pca.transform(scaled_data1)
print(scaled_data1.shape)
print(X.shape)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:, 0],X[:, 1],X[:, 2], c = cancer['target'],cmap = 'plasma')
ax.set_xlabel('First Component')
ax.set_ylabel('Second Component')
ax.set_zlabel('Third Component')
plt.show()
```

```python
pca.components_
```

->In this numpy array each row represent a principal componet, and each column realte back to the original features

->we can visualize this relationship with a heatmap

```python
df_comp = pd.DataFrame(pca.components_,columns = cancer['feature_names'])
plt.figure(figsize =(12,6))
sns.heatmap(df_comp,cmap = 'plasma')
```
