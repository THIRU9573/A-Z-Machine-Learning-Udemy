# what  is Uppper Confidence bound :

Upper Confidence Bound (UCB) is a popular algorithm used in multi-armed bandit problems, where a player has to decide which arm of a slot machine to pull in order to maximize their rewards. The algorithm balances the exploitation of the arms that have already shown to yield high rewards, and the exploration of the unexplored arms, to identify the best one. Here are the steps of UCB algorithm with an example:

Step 1: Initialization

At the start, UCB algorithm initializes the rewards and count of each arm to zero, and the total number of rounds (or pulls) to zero.

Step 2: Calculate UCB for each arm
 
The algorithm calculates the Upper Confidence Bound for each arm using the following formula:

UCB = avg_reward + sqrt(2 * log(total_rounds) / arm_count)

where,

avg_reward: the average reward obtained from that arm
total_rounds: the total number of rounds (or pulls) so far
arm_count: the number of times that arm has been pulled
The formula has two terms: the first term is the average reward from that arm, which indicates exploitation, and the second term is a measure of the uncertainty around the average reward, which indicates exploration.

Step 3: Choose arm with highest UCB

The algorithm chooses the arm with the highest UCB value as the next arm to pull.

Step 4: Pull the chosen arm and update counts and rewards

The algorithm pulls the chosen arm and receives a reward. It updates the count and total reward for that arm, and also updates the total number of rounds.

Step 5: Repeat

The algorithm repeats the steps from 2 to 4 until a predefined number of rounds or a stopping criterion is reached.

Example:
Let's say we have a slot machine with 4 arms, and we want to use UCB algorithm to identify the best arm to maximize the reward. The table below shows the initial state of the algorithm:

Arm	Reward	Count	UCB
1	   0	  0 	0
2      0	  0 	0
3	   0	  0 	0
4	   0	  0 	0
Total	0	0	-
The algorithm starts with zero reward and zero count for each arm, and a total count of zero.

Assuming the algorithm starts with the first arm, and the reward obtained is 5. Then the algorithm calculates the UCB for each arm as follows:

UCB_arm1 = 5/1 + sqrt(2 * log(1) / 1) = 5 + 0 = 5
UCB_arm2 = 0/0 + sqrt(2 * log(1) / 0) = undefined (division by zero)
UCB_arm3 = 0/0 + sqrt(2 * log(1) / 0) = undefined (division by zero)
UCB_arm4 = 0/0 + sqrt(2 * log(1) / 0) = undefined (division by zero)

Since the UCB for arms 2, 3, and 4 are undefined, the algorithm chooses arm 1 as the next arm to pull. The algorithm pulls the arm and updates the count and reward for arm 1:

Arm	Reward	Count	UCB
1	   5	   1	5
2	   0	   0	undefined
3	   0	   0	undefined
4	   0	   0

# Import Libraries :

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

# Importing the daset:

```python
data = pd.read_csv("C:/Users/thiru/Downloads/Resume/Machine Learning A-Z (Codes and Datasets)/Part 6 - Reinforcement Learning/Section 32 - Upper Confidence Bound (UCB)/Python/Ads_CTR_Optimisation.csv")
```

```python
data.head()
```

# implenting the UCB ON DATSET:

```python
N = 10000
d = 10
ads_selected = []
numbers_of_selections = [0]*d
sums_of_rewards = [0]*d
total_reward = 0
for n in range(0, N):
    ad = 0
    max_upper_bound = 0
    for i in range(0, d):
        if (numbers_of_selections[i] > 0):
            average_reward =  sums_of_rewards[i]/numbers_of_selections[i]
            delta_i = math.sqrt(3/2*math.log(n+1)/numbers_of_selections[i])
            upper_bound = average_reward + delta_i
        else:
            upper_bound = 1e400
        if (upper_bound > max_upper_bound):
            max_upper_bound = upper_bound
            ad = i
    ads_selected.append(ad)
    numbers_of_selections[ad] = numbers_of_selections[ad]+1
    reward = data.values[n,ad]
    sums_of_rewards[ad] = sums_of_rewards[ad] + reward
    total_reward = total_reward + reward
        
        
            
            
            
    
    
```

# Visualising the result:

```python
plt.hist(ads_selected)
plt.title('Histograme of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
